```python
# Get the mean , median , quartiles of each numerical variable within this dataset.
df.describe()
```

```python
#Spot the possible outliers in your dataset and get rid of them.
# 1. Calculate Zscore 
for col in df.describe().columns:
    col_zscore = col +'_zscore'
    df[col_zscore] = (df[col] - df[col].mean())/df[col].std()
```

```python
# The outliers in last col_zscore 
df.loc[df[col_zscore]>3]
```

```python
# 3. Get rid of the outliers 
N = 5
# Select last 5 columns of dataframe
last_n_column  = df.iloc[: , -N:]
for col in last_n_column :
    df = df[abs(df[col])<=3] # remove the outliers 
```

```python
# Verification
df.loc[df[col]>3] # no outliers left in last col_zscore
```
