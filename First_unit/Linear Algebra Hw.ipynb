```python
# check if diagonal 
def diagonal (m):
    bool = True 
    if (len(m) == len(m[0])):
        for i in range(-len(m)+1,len(m)):
            if i != 0 :
                vector = np.diag(m,i)
                for j in range (len(vector)):
                    if vector[j] != 0 : 
                       bool = False 
                       break 
    else :
        bool = False 
    return bool 

print(diagonal(img)) 
```
